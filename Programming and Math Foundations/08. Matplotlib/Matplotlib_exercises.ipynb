### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('company_sales_data.csv')
plt.plot(df['month_number'],df['total_profit'])
plt.title('Graph')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.show()
```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
fig = plt.figure()
fig = fig.add_axes([0.1, 0.1, 1.3, 1.3])
fig.plot(df['month_number'],df['total_profit'],label='total profit',marker='o', c='r',linestyle='--',linewidth=3)
fig.set_xlabel('Month number')
fig.set_ylabel('Sold units number')
plt.legend(loc='lower right')
plt.title('company sales data of last year')
plt.show()
```

###Â Exercise 10: Read all product sales data and show it using the stack plot.

```python
# Write your code here
x= df['month_number']
y1,y2,y3,y4,y5,y6 = df['facecream'],df['facewash'],df['toothpaste'],df['bathingsoap'],df['shampoo'],df['moisturizer']
fig,ax = plt.subplots(figsize=(12,8))
ax.set_xlabel('number of month')
ax.set_ylabel('total sales')
ax.set_title('all products sales')
color_map = ['#FF1493','#7AC5CD','#FF4040','#000000','#458B00','#EEC900']

ax.stackplot(x,y1,y2,y3,y4,y5,y6,labels =df.columns[1:-2],colors=color_map)
plt.legend(loc='upper left')
```
